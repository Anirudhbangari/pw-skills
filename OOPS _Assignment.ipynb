```python
#15. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide examples using operators like `+` and `*`.
'''Operator overloading in Python refers to the ability to define special behaviors for the standard operators, such as +, -, *, /, etc., 
for user-defined objects. By implementing certain methods in a class, you can customize how instances of that class respond to operators. 
This concept is closely related to polymorphism as it allows objects of different types to interact with operators in a way that is meaningful for 
those types.

In Python, operator overloading is achieved by defining special methods with double underscores (also known as "dunder" methods or magic methods).
For example, the __add__ method is called when the + operator is used on instances of a class.
'''
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the addition operator
    def __add__(self, other):
        if isinstance(other, Point):
            return Point(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Unsupported operand type for +: " + str(type(other)))

    # Overloading the multiplication operator
    def __mul__(self, scalar):
        if isinstance(scalar, (int, float)):
            return Point(self.x * scalar, self.y * scalar)
        else:
            raise TypeError("Unsupported operand type for *: " + str(type(scalar)))

    def __str__(self):
        return f"Point({self.x}, {self.y})"

# Usage
point1 = Point(1, 2)
point2 = Point(3, 4)

# Overloading addition operator (+)
result_addition = point1 + point2
print(result_addition)  # Output: Point(4, 6)

# Overloading multiplication operator (*)
result_multiplication = point1 * 2
print(result_multiplication)  # Output: Point(2, 4)
'''how this relates to polymorphism:

1.Polymorphic Behavior:
By overloading operators, instances of the Point class can interact with these operators in a way that is specific to their type. 
For example, the + operator concatenates points, and the * operator scales a point by a scalar.
The same operators are used, but their behavior is polymorphic, adapting to the specific type of objects involved in the operation.

2.Code Flexibility:
Operator overloading increases code flexibility by allowing custom types to participate in standard operations. 
This flexibility contributes to a more natural and expressive coding style.

3.Readability:
The use of overloaded operators can make code more readable and intuitive, as it allows you to express operations in a way 
that is conceptually natural for the objects involved.
'''
```

```python
#17. Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) and implement polymorphism through a 
#common `calculate_salary()` method. 
class Employee:
    def __init__(self, name, role):
        self.name = name
        self.role = role

    def calculate_salary(self):
        raise NotImplementedError("Subclasses must implement the calculate_salary method.")

    def __str__(self):
        return f"{self.role}: {self.name}"

class Manager(Employee):
    def __init__(self, name, team_size):
        super().__init__(name, role="Manager")
        self.team_size = team_size

    def calculate_salary(self):
        base_salary = 60000
        bonus_per_employee = 2000
        total_salary = base_salary + (self.team_size * bonus_per_employee)
        return total_salary

class Developer(Employee):
    def __init__(self, name, programming_language):
        super().__init__(name, role="Developer")
        self.programming_language = programming_language

    def calculate_salary(self):
        base_salary = 70000
        language_bonus = 3000
        total_salary = base_salary + language_bonus
        return total_salary

class Designer(Employee):
    def __init__(self, name, years_of_experience):
        super().__init__(name, role="Designer")
        self.years_of_experience = years_of_experience

    def calculate_salary(self):
        base_salary = 65000
        experience_bonus = 1000 * self.years_of_experience
        total_salary = base_salary + experience_bonus
        return total_salary

# Example usage:

# Creating instances of different employee types
manager = Manager("Alice", team_size=10)
developer = Developer("Bob", programming_language="Python")
designer = Designer("Charlie", years_of_experience=5)

# Calculating and displaying salaries using polymorphism
employees = [manager, developer, designer]

for employee in employees:
    print(f"{str(employee)} - Salary: ${employee.calculate_salary():,.2f}")
```

```python
#4. Explain the concept of abstract classes in Python and how they are defined using the `abc` module. Provide an example.
'''In Python, an abstract class is a class that cannot be instantiated and may contain abstract methods—methods without a defined implementation. Abstract classes serve as blueprints for other classes and provide a common interface for a group of related classes. Python's abc module (Abstract Base Classes) is used to define abstract classes and abstract methods.

Here's an explanation of abstract classes in Python and an example using the abc module:

Defining Abstract Classes:

1.Abstract Base Class (ABC):
To define an abstract class in Python, you typically use the ABC metaclass provided by the abc module.
The ABC metaclass is used as a base class for your abstract class.

2.Abstract Method:
An abstract method is a method declared in an abstract class without providing an implementation.
Subclasses are required to provide a concrete implementation for each abstract method.
for example:'''

from abc import ABC, abstractmethod

class shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
class circle(shape):
    def __init__(self,radius):
        self.radius = radius
        
    def area(self):
        return f" Circle : {22*(self.radius**2)/7}"
    
class square(shape):
    def __init__(self,side):
        self.side = side
        
    def area(self):
        return f"Square : {self.side**2}"

class triangle(shape):
    def __init__(self,si,sid):
        self.side = si
        self.side1 = sid
    
    def area(self):
        return f"Triangle  {self.side*self.side1*0.5}"
    
ci = circle(7)
sq = square(8)
tr = triangle(4,8)

print("Area",ci.area())
print("Area",sq.area())
print("Area",tr.area())
```

```python
#13. Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) and implement abstraction by defining a common 
#'get_salary()' method.
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, role):
        self.name = name
        self.role = role

    @abstractmethod
    def get_salary(self):
        pass

    def __str__(self):
        return f"{type(self).__name__}: {self.name}"

class Manager(Employee):
    def __init__(self, name, team_size):
        super().__init__(name, role="Manager")
        self.team_size = team_size

    def get_salary(self):
        base_salary = 80000
        bonus_per_employee = 2500
        total_salary = base_salary + (self.team_size * bonus_per_employee)
        return total_salary

class Developer(Employee):
    def __init__(self, name, programming_language):
        super().__init__(name, role="Developer")
        self.programming_language = programming_language

    def get_salary(self):
        base_salary = 90000
        language_bonus = 4000
        total_salary = base_salary + language_bonus
        return total_salary

class Designer(Employee):
    def __init__(self, name, years_of_experience):
        super().__init__(name, role="Designer")
        self.years_of_experience = years_of_experience

    def get_salary(self):
        base_salary = 85000
        experience_bonus = 1200 * self.years_of_experience
        total_salary = base_salary + experience_bonus
        return total_salary

# Creating instances of different employee types
manager = Manager("Alice", team_size=10)
developer = Developer("Bob", programming_language="Python")
designer = Designer("Charlie", years_of_experience=5)

# Retrieving and displaying salaries using abstraction
employees = [manager, developer, designer]

for employee in employees:
    print(f"{str(employee)} - Salary: ${employee.get_salary():,.2f}")
```

```python
#16. Create a Python class for a computer system, demonstrating abstraction by defining common methods (e.g. 'power_on()','shutdown()') 
#in abstract baseclass
from abc import ABC, abstractmethod

class Computer(ABC):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.powered_on = False

    @abstractmethod
    def power_on(self):
        pass

    @abstractmethod
    def shutdown(self):
        pass

    def __str__(self):
        return f"{self.brand} {self.model}"

class DesktopComputer(Computer):
    def __init__(self, brand, model, monitor_type):
        super().__init__(brand, model)
        self.monitor_type = monitor_type

    def power_on(self):
        if not self.powered_on:
            print(f"{str(self)} is powering on.")
            self.powered_on = True
        else:
            print(f"{str(self)} is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{str(self)} is shutting down.")
            self.powered_on = False
        else:
            print(f"{str(self)} is already powered off.")

class LaptopComputer(Computer):
    def __init__(self, brand, model, battery_life):
        super().__init__(brand, model)
        self.battery_life = battery_life

    def power_on(self):
        if not self.powered_on:
            print(f"{str(self)} is powering on. Battery life: {self.battery_life}%.")
            self.powered_on = True
        else:
            print(f"{str(self)} is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{str(self)} is shutting down.")
            self.powered_on = False
        else:
            print(f"{str(self)} is already powered off.")

# Creating instances of different computer types
desktop = DesktopComputer("Dell", "OptiPlex", "LED")
laptop = LaptopComputer("HP", "Envy", battery_life=80)

# Performing common actions using abstraction
computers = [desktop, laptop]

for computer in computers:
    print(f"{str(computer)}:")
    computer.power_on()
    computer.shutdown()
    print()
```

```python
#11. How can you encapsulate and hide the details of composed objects in Python classes to maintain abstraction?
'''To encapsulate and hide the details of composed objects in Python classes and maintain abstraction, 
you can use a combination of encapsulation principles, such as private attributes, and provide a well-defined interface for interacting with the 
composed objects. Here's a step-by-step guide:

1.Define an Interface or Abstract Class:
Create an interface or an abstract class that declares the methods expected from the composed objects.

2.Implement Concrete Components:
Implement concrete classes that represent the components. Each concrete class should inherit from the interface or abstract class.

3.Create a Composed Object Class:
Create a class that will act as the composed object. This class should encapsulate the details of the composed objects and 
provide a clean interface for interaction.

for example: '''
from abc import ABC, abstractmethod

class shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
class circle(shape):
    def __init__(self,radius):
        self.radius = radius
        
    def area(self):
        return f" Circle : {22*(self.radius**2)/7}"
    
class square(shape):
    def __init__(self,side):
        self.side = side
        
    def area(self):
        return f"Square : {self.side**2}"

class triangle(shape):
    def __init__(self,si,sid):
        self.side = si
        self.side1 = sid
    
    def area(self):
        return f"Triangle  {self.side*self.side1*0.5}"
    
ci = circle(7)
sq = square(8)
tr = triangle(4,8)

print("Area",ci.area())
print("Area",sq.area())
print("Area",tr.area())
```

```python
#14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.
class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    def __str__(self):
        return f"{self.quantity} {self.unit} of {self.name}"


class Dish:
    def __init__(self, name, description, price, ingredients):
        self.name = name
        self.description = description
        self.price = price
        self.ingredients = ingredients

    def display_dish(self):
        print(f"Name: {self.name}")
        print(f"Description: {self.description}")
        print(f"Price: ${self.price:.2f}")
        print("Ingredients:")
        for ingredient in self.ingredients:
            print(f"  - {ingredient}")

    def __str__(self):
        return f"Dish: {self.name}"


class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes

    def display_menu(self):
        print(f"Menu: {self.name}")
        for dish in self.dishes:
            dish.display_dish()
            print("\n")

    def __str__(self):
        return f"Menu: {self.name}"


class Restaurant:
    def __init__(self, name, menus):
        self.name = name
        self.menus = menus

    def display_restaurant(self):
        print(f"Restaurant: {self.name}")
        for menu in self.menus:
            menu.display_menu()

    def __str__(self):
        return f"Restaurant: {self.name}"

# Create ingredients
tomato = Ingredient("Tomato", 2, "pieces")
cheese = Ingredient("Cheese", 150, "grams")
dough = Ingredient("Dough", 1, "piece")

# Create dishes
margherita_pizza = Dish("Margherita Pizza", "Classic pizza with tomato and cheese", 10.99, [tomato, cheese, dough])
spaghetti_bolognese = Dish("Spaghetti Bolognese", "Spaghetti with meat sauce", 12.99, [tomato, cheese])

# Create menus
lunch_menu = Menu("Lunch Menu", [margherita_pizza, spaghetti_bolognese])

# Create a restaurant
my_restaurant = Restaurant("Italian Delight", [lunch_menu])

# Display restaurant information
my_restaurant.display_restaurant()
```

```python
#16. Create a Python class for a computer game character, using composition to represent attributes like weapons, armor, and inventory.
class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def __str__(self):
        return f"Weapon: {self.name} (Damage: {self.damage})"


class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def __str__(self):
        return f"Armor: {self.name} (Defense: {self.defense})"


class Inventory:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def display_inventory(self):
        print("Inventory:")
        for item in self.items:
            print("-", item)


class GameCharacter:
    def __init__(self, name, health):
        self.name = name
        self.health = health
        self.weapon = None
        self.armor = None
        self.inventory = Inventory()

    def equip_weapon(self, weapon):
        self.weapon = weapon

    def equip_armor(self, armor):
        self.armor = armor

    def display_character_info(self):
        print(f"Character: {self.name}")
        print(f"Health: {self.health}")
        if self.weapon:
            print(f"Equipped {self.weapon}")
        if self.armor:
            print(f"Equipped {self.armor}")
        self.inventory.display_inventory()

# Create weapons, armor, and a character
sword = Weapon("Sword", 20)
shield = Armor("Shield", 10)
character = GameCharacter("Hero", 100)

# Equip weapons and armor
character.equip_weapon(sword)
character.equip_armor(shield)

# Add items to the inventory
health_potion = "Health Potion"
character.inventory.add_item(health_potion)

# Display character information
character.display_character_info()
```

```python
#17. Describe the concept of "aggregation" in composition and how it differs from simple composition.
'''In object-oriented programming, "aggregation" is a form of composition where one object is composed of multiple other objects, 
but the relationship is more relaxed than in simple composition. Aggregation implies a "has-a" relationship between the containing object (the whole) 
and its parts. The key characteristic of aggregation is that the parts can exist independently of the whole.

*******Key Characteristics of Aggregation:******

1.Independence of Lifetimes:
In aggregation, the objects representing the parts have an independent existence and can exist even if the containing object is destroyed.

2.Weaker Relationship:
The relationship between the containing object and its parts is considered weaker compared to simple composition. The parts can be shared among multiple containing objects.

3.Multiplicity:
Multiplicity allows for zero or more instances of the parts to be associated with a single instance of the whole. 
This flexibility distinguishes aggregation from stronger forms of composition.

for example:'''
class Department:
    def __init__(self, name):
        self.name = name

class University:
    def __init__(self, name):
        self.name = name
        self.departments = []  # Aggregation

    def add_department(self, department):
        self.departments.append(department)

# Creating instances
math_department = Department("Mathematics")
physics_department = Department("Physics")

my_university = University("ABC University")
my_university.add_department(math_department)
my_university.add_department(physics_department)


'''Differences from Simple Composition:

>Independence:
In simple composition, the parts are typically created and managed by the containing object. In aggregation, 
the parts can exist independently and may be created elsewhere.

>Multiplicity:
Aggregation allows for a flexible multiplicity, where there can be zero or more instances of the parts associated with a single instance of the whole. 
Simple composition implies a one-to-one or one-to-many relationship.

>Ownership:
In simple composition, the containing object usually has a strong ownership and responsibility for the creation and destruction of its parts. 
In aggregation, the parts can be shared among multiple containing objects.

>Lifecycle Independence:
Aggregated objects have a lifecycle independent of the containing object. If the containing object is destroyed, 
the aggregated objects may continue to exist.

****Visual Representation:
In UML (Unified Modeling Language), aggregation is often represented by an open diamond shape at the end of the association line between 
the containing class and the contained class.

+---------------------+
|     University     |
|---------------------|
| - name: str         |
| - departments: list |
|---------------------|
| + add_department()  |
+---------------------+
          |
          |
         / \
        /   \
+---------------------+
|     Department     |
|---------------------|
| - name: str         |
+---------------------+

The open diamond indicates an aggregation relationship.'''
```
